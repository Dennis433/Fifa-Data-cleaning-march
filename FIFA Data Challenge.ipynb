### Cleaning the Wage Column

```python
fifa_data["Wage"].unique()
```

As we can see In the above unique values under this column, some players earn in hundreds  while some in thousand of euro

We have to be careful while Slicing 

```python
fifa_data['Wage'] = fifa_data['Wage'].str.replace('€','')
```

```python
# Function will clean Column Wage
def convert_wage(val):
    
    if "K" in val:
        value=val.split("K")[0]
        value= int(value)*1000
        return value
    else:
        value= int(val)

        return val
```

```python
fifa_data['Wage']=fifa_data['Wage'].apply(convert_wage)
```

```python
fifa_data['Wage'].unique()
```

```python
### Converting it to the correct datatype(integer)
fifa_data['Wage'] = pd.to_numeric(fifa_data['Wage'])
```

```python
#Confirming there are no null values
fifa_data["Wage"].isna().sum()
```

#### Cleaning the value column

```python
fifa_data['Value'].unique()
```

The above column has almost the same issues as the above column but in a slightly different way


```python
fifa_data['Value'] = fifa_data['Value'].str.replace('€','')
```

```python
def convert_value(val):
    
    if 'M' in val:
        # We split val by space and then extract using the index of the vl
        x=float(val.split('M')[0])
        # After extracting number fron val, we multiply by 1000000
        return int(x*1000000)

    elif 'K' in val:
        x=float(val.split('K')[0])
        # After extracting number fron val, we multiply by 1000000
        return int(x*1000)
    else:
        return int(val)
```

```python
#Applying the above created function on the value column
fifa_data['Value'] = fifa_data['Value'].apply(convert_value)
```

```python
fifa_data['Value'].unique()
```

```python
##confirming there are no null values
fifa_data["Value"].isna().sum()
```

### Dealing with the contract column

```python
fifa_data["Contract"].unique()
```

In the above column it can be seen that some players were bought while some were free and others were on loan it would be better to create a new column for categorizing them

```python
def player_status(val):
    val = val.split(" ")
    if 'Free' in val:
        return "Free"
    elif "Loan" in val:
        return "Loan"
    elif "~" in val:
        return 'Contract'
```

### Using the function above to apply on the new column

```python
fifa_data['Status'] = fifa_data['Contract'].apply(player_status)
```

```python
fifa_data['Status'].unique()
```

#### Creating New Column "Contract_end"

```python
# Function create new column "Contract_end" for every player 
def convert_contract(val):
    # We split values into list by space
    val= val.split(" ")
    if 'Free' in val:
        return val[0]
    elif "Loan" in val:
        return val[2]
    elif "~" in val:
        return val[2]
```

```python
fifa_data['end_year'] = fifa_data["Contract"].apply(convert_contract)
```

Removing star symbol from column W/F,SM and IP

```python
fifa_data.loc[:,["SM","IR",'W/F']]
```

```python
# Function to remove star
def star_remove(val):
    s = [int(s) for s in re.findall(r'[0-9]+', val)]
    return s[0]
```

Applying function to the necessary columns

```python
fifa_data['W/F'] = fifa_data['W/F'].apply(star_remove)
fifa_data['IR']  = fifa_data['IR'].apply(star_remove)
fifa_data['SM']  = fifa_data['SM'].apply(star_remove)
```
